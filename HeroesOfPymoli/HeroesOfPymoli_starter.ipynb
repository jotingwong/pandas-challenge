```python
# Author: JoTing Wong

# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)

```

## Player Count

* Display the total number of players


```python
TotalPlayers = len(purchase_data["SN"].unique())

TotalPlayers_df = pd.DataFrame({"Total Players":[TotalPlayers]})

TotalPlayers_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
UniqueItems = len(purchase_data["Item ID"].unique())
AvgPrice = purchase_data["Price"].mean()
NumofPurchase = len(purchase_data["Purchase ID"])
TotalRevenue = purchase_data["Price"].sum()

Summary_df = pd.DataFrame({"Number of Unique Items":[UniqueItems],
                            "Average Price" : [AvgPrice],
                            "Number of Purchases" : [NumofPurchase],
                            "Total Revenue" : [TotalRevenue]})

Summary_df["Average Price"] = Summary_df["Average Price"].map("${:,.2f}".format)
Summary_df["Total Revenue"] = Summary_df["Total Revenue"].map("${:,.2f}".format)

Summary_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python

gender_group = purchase_data.groupby("Gender").nunique()
gender_group["Total Players"] = TotalPlayers
gender_group["Percentage of Players"] = ((gender_group["SN"]/gender_group["Total Players"])*100).map("{:,.2f}%".format)

# remove unwanted columns
gender_df = gender_group.drop(["Purchase ID","Age","Gender","Item ID","Item Name","Price","Total Players"], axis=1)

# rename column
gender_df = gender_df.rename(columns = {"SN": "Total Count"})

gender_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
gender_purchase_group = purchase_data.groupby("Gender")

Total_Price = gender_purchase_group["Price"].sum()
Total_Purchase = gender_purchase_group["Purchase ID"].count()
Gender_Count = gender_purchase_group["SN"].nunique()

Avg_Price = Total_Price/Total_Purchase
Price_per_Person = Total_Price/Gender_Count


# create dataframe for output
Purchase_df = pd.DataFrame({"Purchase Count": Total_Purchase,
                            "Average Purchase Price" : Avg_Price,
                            "Total Purchase Value": Total_Price,
                            "Average Total Purchase per Person": Price_per_Person 
                           })

# format columns 
Purchase_df["Average Purchase Price"] = Purchase_df["Average Purchase Price"].map("${:,.2f}".format)
Purchase_df["Average Total Purchase per Person"] = Purchase_df["Average Total Purchase per Person"].map("${:,.2f}".format)
Purchase_df["Total Purchase Value"] = Purchase_df["Total Purchase Value"].map("${:,.2f}".format)


Purchase_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python

bins = [0, 9, 14, 19, 24, 29, 34, 39, 50]

# Create the names for the four bins
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

purchase_data["Age Bucket"] = pd.cut(purchase_data["Age"],bins,labels = group_names)

age_bucket_group = purchase_data.groupby("Age Bucket")

total_count = age_bucket_group["SN"].nunique()
pct_player = (total_count/TotalPlayers) * 100

# create dataframe for output
age_df = pd.DataFrame({"Total Count": total_count,
                            "Percentage of Players" : pct_player
                           })

# format columns 
age_df["Percentage of Players"] = age_df["Percentage of Players"].map("{:,.2f}%".format)


age_df

```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python

purchase_count = age_bucket_group["Age"].count()
total_price_value = age_bucket_group["Price"].sum()
total_age_count = age_bucket_group["SN"].nunique()

avg_purchase_price = total_price_value/purchase_count
avg_purchase_per_person = total_price_value/total_age_count


age_purchase_df = pd.DataFrame({"Purchase Count": purchase_count,
                                "Average Purchse Price": avg_purchase_price,
                                "Total Price Value" : total_price_value,
                                "Average Total Purchase per Person": avg_purchase_per_person
                               })

age_purchase_df["Average Purchse Price"] = age_purchase_df["Average Purchse Price"].map("${:,.2f}".format)
age_purchase_df["Total Price Value"] = age_purchase_df["Total Price Value"].map("${:,.2f}".format)
age_purchase_df["Average Total Purchase per Person"] = age_purchase_df["Average Total Purchase per Person"].map("${:,.2f}".format)


age_purchase_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
top_spender_group = purchase_data.groupby("SN")

spender_price = top_spender_group["Price"].sum()
spender_purchase_count = top_spender_group["Purchase ID"].count()

spender_avg_price = spender_price/spender_purchase_count


# create dataframe for output
top_spender_df = pd.DataFrame({"Purchase Count": spender_purchase_count,
                               "Average Purchase Price": spender_avg_price,
                               "Total Purchase Value": spender_price
                              })

top_spender_df = top_spender_df.sort_values(by="Total Purchase Value", ascending=False)

# # format columns 
top_spender_df["Average Purchase Price"] = top_spender_df["Average Purchase Price"].map("${:,.2f}".format)
top_spender_df["Total Purchase Value"] = top_spender_df["Total Purchase Value"].map("${:,.2f}".format)


top_spender_df
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
item_group = purchase_data.groupby(["Item ID","Item Name"])

item_total_price = item_group["Price"].sum()
item_count = item_group["Purchase ID"].count()
item_price = item_group["Price"].max()



# create dataframe for output
popular_item_df = pd.DataFrame({"Purchase Count": item_count,
                               "Item Price": item_price,
                               "Total Purchase Value": item_total_price
                              })

popular_item_df = popular_item_df.sort_values(by="Purchase Count", ascending=False)

# # format columns 
popular_item_df["Item Price"] = popular_item_df["Item Price"].map("${:,.2f}".format)
popular_item_df["Total Purchase Value"] = popular_item_df["Total Purchase Value"].map("${:,.2f}".format)


popular_item_df
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
# create dataframe for output
most_profit_df = pd.DataFrame({"Purchase Count": item_count,
                               "Item Price": item_price,
                               "Total Purchase Value": item_total_price
                              })



most_profit_df = most_profit_df.sort_values(by="Total Purchase Value", ascending=False)

# format columns 
most_profit_df["Item Price"] = most_profit_df["Item Price"].map("${:,.2f}".format)
most_profit_df["Total Purchase Value"] = most_profit_df["Total Purchase Value"].map("${:,.2f}".format)


most_profit_df
```
